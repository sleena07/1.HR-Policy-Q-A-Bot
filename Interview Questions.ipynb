### What is a system prompt and why does it matter?

```python
It's the instructions given to the model before the conversation starts. 
It controls the model's behaviour, persona, and constraints. 
Without a good system prompt the model answers anything without restriction.
```

### What is session state in Streamlit and why do you need it?

```python
Streamlit reruns the entire script on every interaction. 
Session state persists data across reruns. 
Without it the chat history would reset every time someone typed a message.
```

### Why did you pass chat history to the API?

```python
LLMs have no memory between API calls. Each call is stateless. 
Passing history gives the model context of the conversation so it can give coherent follow-up answers.
```

### What is max_tokens?

```python
Controls the maximum length of the response. Important for cost control — longer responses cost more.
```

### How did you prevent the model from making up answers?

```python
Through the system prompt rules that explicitly say to answer only from the policy document 
and to admit when information isn't available.
```

### Walk me through what happens from when a user types a question to when they see the answer



```python
User types question. Streamlit captures it in typed_input. 
It gets appended to session state chat history. 
The ask_hr_bot function is called with the question and previous history. 
The function builds the messages array from history. 
It calls the OpenAI API with a system prompt containing the HR policy and the messages array. 
OpenAI processes everything and returns a response object. The function extracts the text content from the response object. 
Streamlit displays it in an assistant chat bubble and saves it to session state for future context.
```
